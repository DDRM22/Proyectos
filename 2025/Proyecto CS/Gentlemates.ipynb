```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import awpy
import streamlit as st # Investigar como montarlo en local
```

```python
# Cargamos demos de los partidos
from awpy.demo import Demo

demo_m8_vs_g2_inferno = Demo("data/input/Demos/Gentlemates/Inferno/g2-vs-gentle-mates-m1-inferno.dem", verbose=True)
demo_m8_vs_g2_inferno.parse()  # Parseamos el demo

demo_m8_vs_g2_ancient = Demo("data/input/Demos/Gentlemates/Ancient/g2-vs-gentle-mates-m3-ancient.dem", verbose=True)
demo_m8_vs_g2_ancient.parse()  # Parseamos el demo
```

```python
# Analizamos los datos de la demo cargada

print(f"Encabezado: \n{demo_m8_vs_g2_inferno.header}\n")
print(f"Rondas: \n{demo_m8_vs_g2_inferno.rounds.head(n=3)}\n") # .head(n=3) para ver solo las primeras 3 rondas
print(f"Eventos: \n{demo_m8_vs_g2_inferno.events}\n")
print(f"Granadas: \n{demo_m8_vs_g2_inferno.grenades}\n")
print(f"Daños: \n{demo_m8_vs_g2_inferno.damages}\n")

```

```python
# Extrayendo datos específicos de la demo

from awpy.stats import adr, kast, impact, rating
print("ADR:", adr(demo_m8_vs_g2_inferno))
print("KAST:", kast(demo_m8_vs_g2_inferno))
print("Impact:", impact(demo_m8_vs_g2_inferno))
print("Rating:", rating(demo_m8_vs_g2_inferno))
print("ADR:", adr(demo_m8_vs_g2_inferno))
```
